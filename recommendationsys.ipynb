# **Recommendation**

```python
from transformers import BertModel, BertTokenizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import label_ranking_average_precision_score, ndcg_score

import torch

# Define a function to prepare the data and load the model
def prepare_data(condition, features, Drugdf):
    # Create a new column with the combined features
    Drugdf['combined_features'] = Drugdf[features].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

    # Filter the dataset by the input condition
    df_condition = Drugdf[Drugdf['Condition'] == condition]

    # Load pre-trained model and tokenizer
    bert_model = BertModel.from_pretrained('bert-base-uncased')
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    return df_condition, bert_model, tokenizer

# Define a function to generate BERT embeddings for combined_features
def get_embeddings(df, bert_model, tokenizer):
    embeddings = []
    for text in df['combined_features']:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
        with torch.no_grad():
            outputs = bert_model(**inputs)
        embeddings.append(outputs.last_hidden_state.mean(dim=1).squeeze())
    embeddings = torch.stack(embeddings)
    return embeddings

# Define a function to print the recommended drugs and the evaluation metrics based on cosine similarity
def print_recommendations(condition, embeddings, df):
    # Calculate the cosine similarity between the documents
    cosine_sim = cosine_similarity(embeddings, embeddings)

    # Get the indices of the top 5 drugs with the highest cosine similarity
    drug_indices = np.argsort(cosine_sim[-1])[:-11:-1]

    # Print the recommended drugs
    print('Recommended drugs for {}:'.format(condition))
    for idx in drug_indices[1:]:
        print('- {} ({}): {}€'.format(df.iloc[idx]['Drug'], df.iloc[idx]['Form'], df.iloc[idx]['Price']))
        
    # Convert the Satisfaction column into binary labels
    df['Satisfaction'] = df['Satisfaction'].apply(lambda x: 1 if x >= 0.5 else 0)

    # Get the ground truth relevance labels for the drugs
    y_true = df['Satisfaction'].values.reshape(1, -1)

    # Get the predicted relevance scores for the drugs based on cosine similarity
    y_score = cosine_sim[-1].reshape(1, -1)

    # Calculate the MAP and NDCG scores
    map_scr = label_ranking_average_precision_score(y_true, y_score)
    ndcg_scr = ndcg_score(y_true, y_score)

    # Print the evaluation metrics
    print('MAP score: {:.4f}'.format(map_scr))
    print('NDCG score: {:.4f}'.format(ndcg_scr))
```
